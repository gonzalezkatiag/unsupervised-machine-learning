```python
import pandas as pd

import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
```

```python
file_path = Path("crypto_data.csv")
crypto_df = pd.read_csv(file_path, index_col = 0)
crypto_df.head(5)
```

```python
# Only keep cryptocurrencies that are trading
crypto_df = crypto_df.loc[crypto_df['IsTrading'] == True]
crypto_df
```

```python
# Delete 'IsTrading' column
crypto_df = crypto_df.drop(columns = 'IsTrading')
crypto_df.head()
```

```python
# Removing rows w/ at least 1 null val
crypto_df = crypto_df.dropna()
crypto_df.isnull().sum()
```

```python
# Removing rows w/ cryptocurrencies that have zero coins mined
crypto_df = crypto_df.loc[crypto_df['TotalCoinsMined'] != 0]
crypto_df
```

```python
# Storing 'CoinName' in its own DF
cryptoname_df = pd.DataFrame(data = crypto_df, columns = ['CoinName'])
cryptoname_df
```

```python
# Droppin 'CoinName' from crypto df
crypto_df = crypto_df.drop(columns = 'CoinName')
crypto_df
```

```python
# Creating dummy variables 
crypto = pd.get_dummies(crypto_df, columns=['Algorithm', 'ProofType'])
crypto
```

```python
# Standardizing data
crypto_scaled = StandardScaler().fit_transform(crypto)
crypto_scaled
```

```python
# Initialize PCA model & getting 3 principal components
pca = PCA(n_components=0.90)

crypto_pca = pca.fit_transform(crypto_scaled)
```

```python
# Transforming PCA data to a DF
crypto_pca_df = pd.DataFrame(data=crypto_pca)
crypto_pca_df.head()
```

```python
# Reducing dimensions of pricipal components w/ t-SNE
tsne = TSNE(learning_rate=30)
tsne_features = tsne.fit_transform(crypto_pca_df)
tsne_features
```

```python
sns.scatterplot(x=tsne_features.T[0], y=tsne_features.T[1])
plt.title('t-SNE Scatter Plot')
```

```python
inertia = []
k = list(range(1, 11))

# Calculating inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(crypto_pca_df)
    inertia.append(km.inertia_)

# Creating the Elbow Curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(list(range(11)))
plt.title('Elbow Curve')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

# CONCLUSIONS

Based on the analysis and elbow curve visualization, I would not recommend to cluster the cryptocurrencies as there's no clear k value and therefore no cluster will perform better than the other.
